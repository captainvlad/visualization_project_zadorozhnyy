#### 1. Яка країна має найперспективніше молоде покоління

Для цього візьмемо поле 'potential' із гри FIFA 21 (для гравців не старших за 22 роки), обрахуємо середнє значення потенціалу <br>

для гравця кожної із представлених країн, а потім -- розподілимо отримані дані на мапі.

**Ціль:** Дізнатися у яких країн станом на сьогодні найбільш перспективне (на думку розробників гри) покоління.

**Обраний інструмент:** мапа із кольоровим забарвленням.

**Переваги:** Легко зчитується, показує великий обсяг даних.

**Недоліки:** читач може не знати місцезнаходження потрібної країни на карті.

**Попередня гіпотеза:** найперспективніше молоде покоління мають країни Латинської 

Америки (Аргентина, Бразилія, Уругвай).

```python
df_players_21["pure_potential"] = df_players_21["potential"] - df_players_21["overall"]

df_players_21 = df_players_21.sort_values(by=['pure_potential'], ascending=False)
df_players_21.head()
```

#### 2. Наскільки були правильними припущення розробників FIFA у 2014 році щодо сьогоднішнього дня

Беремо до уваги період часу 2014-2020 роки та гравців, які були не старшими 27 років станом на 2014 рік.

**Ціль:** перевірити правильність оцінки потенціалу, даної розробниками гравцям у розрізі часу (за 6 років).

**Обраний інструмент:** зафарбована гістограма.

**Переваги:** інтерактивність дозволяє легко отримати дані, колір та форма діаграми дозволяють побачити тенденцію.

**Недоліки:** читачу може бути складно зчитати кількість гравців певної колонки без використання тултіпів.

**Попередня гіпотеза:** переважна більшість гравців (~70%) оцінена правильно, а неточність в оцінюванні 

(недооцінка або переоцінка) становить не більше **плюс мінус 15%**

```python
df_15_assumption = df_players_15[['sofifa_id', 'short_name', 'nationality', 'overall', 'potential']]
df_21_reality = df_players_21[['sofifa_id', 'short_name', 'nationality', 'overall', 'potential']]

df_15_assumption = df_15_assumption.loc[(df_players_15['age'] < 27)]
player_15_21_intersection = set(df_15_assumption['sofifa_id']).intersection(set(df_21_reality['sofifa_id']))

df_15_assumption = df_15_assumption.loc[(df_15_assumption['sofifa_id']).isin(player_15_21_intersection)]
df_21_reality = df_21_reality.loc[(df_21_reality['sofifa_id']).isin(player_15_21_intersection)]

df_15_assumption = df_15_assumption.sort_values(by=['sofifa_id'])
df_21_reality = df_21_reality.sort_values(by=['sofifa_id'])
```

```python
df_21_reality['potential_in_2014'] = df_15_assumption['potential'].values

df_21_reality['potential_dynamics'] = df_21_reality['overall'] / df_21_reality['potential_in_2014'] * 100 - 100
df_21_reality['potential_dynamics'] = df_21_reality['potential_dynamics'].astype('int32')
```

```python
def present_potential(value):
    if value > 0:
        return f'+{value}%'
    return f'{value}%'

df_21_reality['potential_dynamics_present'] = df_21_reality['potential_dynamics'].map(lambda x: present_potential(x))
```

```python
alt.Chart(
    df_21_reality,
    title="Правильність припущення розробників FIFA у 2014 році порівняно із сьогоденням"
).mark_bar(
    size=10
).encode(
    x=alt.X(
        'potential_dynamics:Q',
        axis=alt.Axis(
            labelExpr="datum.value > 0 ? '+' + datum.label + '%' : datum.label + '%'"
        ),
        title='Динаміка потенціалу'
    ),
    y=alt.Y(
        'count(potential_dynamics):Q',
        title='Кількість гравців'
    ),
    tooltip = [
        alt.Tooltip('potential_dynamics_present:N', title='Динаміка'),
        alt.Tooltip('count(potential_dynamics)', title='Кількість')
    ],
    color=alt.condition(
        alt.datum.potential_dynamics > 0,
        alt.value("darkgreen"),
        alt.value("darkred")
    )
).properties(
    padding=20,
    width=1200,
    height=800
).configure_title(
    fontSize=24
)
```

**Висновок:** Гіпотеза була правильною, більшість помилок було зроблено в межах 10-15% плюс мінус. 

Також можемо побачити екстремуми, які неможливо передбачити у юному віці.

#### 3. Футболісти яких країн були переоціненими та недооціненими

Для цього використаємо напрацювання попереднього етапу, взявши до уваги показник національності гравця.

Отримані залежності відображаємо на мапі. Беремо до уваги період часу 2014-2020 роки та гравців, які

були не старшими 27 років станом на 2014 рік.

**Ціль:** дізнатися, які країни мають в середньому переоцінених та недооцінених гравців.

**Обраний інструмент:** зафарбована мапа світу.

**Переваги:** легко побачити велику кількість інформації та швидко побачити тенденції.

**Недоліки:** читач може не знати, де знаходиться потрібна країна.

**Попередня гіпотеза:** гравці з країн Східної Європи є найбільш переоціненими у світі.

```python
df_21_reality = df_21_reality[[
    'nationality',
    'potential_dynamics'
]].groupby(['nationality']) \
    .mean() \
    .round(2) \
    .reset_index()

fifa_map_countries = set(df_21_reality['nationality']).intersection(set(world.name))

df_21_reality = df_21_reality.loc[
    (df_21_reality['nationality']).isin(fifa_map_countries)
]

no_data_countries = world.loc[
    ~(world['name'].isin(df_21_reality['nationality']))
]

no_data_countries.insert(3, 'no_data', 'No Data')

data_countries = world.loc[
    (world['name'].isin(df_21_reality['nationality']))
]

data_countries = data_countries.sort_values(by=['name'])
df_21_reality = df_21_reality.sort_values(by=['nationality'])

assert data_countries['name'].to_list() == df_21_reality['nationality'].to_list()

data_countries['potential_dynamics'] = df_21_reality['potential_dynamics'].map(lambda x: x * 100).values
data_countries['potential_dynamics_present'] = df_21_reality['potential_dynamics'].map(lambda x: present_potential(x)).values
```

```python
chart_1 = alt.Chart(
    data_countries,
    title='Найбільш переоцінені та недооцінені країни за потенціалом'
).project(
    type = 'equalEarth'
).mark_geoshape(
    stroke = 'white'
).encode(
    opacity = alt.condition(selection2, alt.value(0.8), alt.value(0.3)),
    color=alt.condition(
        alt.datum.potential_dynamics > 0,
        alt.value("darkgreen"),
        alt.value("darkred")
    ),
    tooltip = [
        alt.Tooltip('name:N', title='Країна'),
        alt.Tooltip('potential_dynamics_present:N', title='Динаміка потенціалу')
    ],
).add_selection(
    selection2
)

chart_2 = alt.Chart(no_data_countries).project(
    type = 'equalEarth'
).mark_geoshape(
    stroke = 'white',
    fill = 'darkgray'
).encode(
    opacity = alt.condition(selection1, alt.value(0.8), alt.value(0.3)),
    tooltip = [
        alt.Tooltip('name:N', title='Країна'),
        alt.Tooltip('no_data:N', title='Динаміка потенціалу')
    ],
).add_selection(
    selection1
).properties(
    width = 1200,
    height = 800
)

alt.layer(
    chart_1,
    chart_2
).properties(
    padding=20,
    width = 1200,
    height = 800
).configure_title(
    fontSize=24
)
```

#### 5. Які номери на спинах є найпопулярнішими серед футболістів

**Ціль:** Дізнатися, як розподіляютья ігрові номери серед футболістів. Які номери 

є найбільш популярними, а які - навпаки?

**Обраний інструмент:** Гістограма.

**Переваги:** легко побачити велику кількість інформації, а за допомогою

тултіпів - легко отримати точні числа.

**Недоліки:** читач може не знати точних чисел, не користуючись тултіпами.

**Попередня гіпотеза:** найбільш популярні номери: 10, 7, 8, 1.


```python
# Clarify jersey data

def define_one_position(i):
    # 'RWB, RB, RM' -> 'RWB'
    if ',' in i:
        coma_index = i.find(',')
        return i[:coma_index]
    else:
        return i

number_mode_df_21 = df_players_21
number_mode_df_21['player_positions'] = df_players_21['player_positions'].map(lambda x: define_one_position(x))

number_mode_df_21.dropna(subset = ['team_jersey_number'], inplace=True)
number_mode_df_21['team_jersey_number'] = number_mode_df_21['team_jersey_number'].astype('int')
```

```python
number_count_df_2020 = pd.DataFrame(columns=['number', 'count_2020'])

def calculate_percentage(year, count):
    percent = count / number_count_df_2020['count_2020'].sum()
    return f'{round(percent * 100, 2)}%'


for number, count in number_mode_df_21['team_jersey_number'].value_counts().iteritems():
    number_count_df_2020 = number_count_df_2020.append({'number': number, 'count_2020': count}, ignore_index=True)

number_count_df_2020['number'] = number_count_df_2020['number'].astype('int')
number_count_df_2020['count_2020'] = number_count_df_2020['count_2020'].astype('int')


number_count_df_2020['percentage'] = number_count_df_2020['count_2020'].map(lambda x: calculate_percentage('2020', x))
number_count_df_2020 = number_count_df_2020.sort_values(by='number')
```

```python
selection3 = alt.selection_single()

alt.Chart(
    number_count_df_2020,
    title="Розподіл гравців за номерами у 2020 році"
).mark_bar().encode(
    opacity = alt.condition(selection3, alt.value(1), alt.value(0.3)),
    x=alt.X('number', title='Номер'),
    y=alt.Y('count_2020:Q', title="Кількість гравців"),
    tooltip = [
        alt.Tooltip('number:N', title='Ігровий номер'),
        alt.Tooltip('count_2020:N', title='Кількість'),
        alt.Tooltip('percentage:N', title='Частка від всіх')
    ]
).add_selection(
    selection3
).properties(
    padding=20,
    width = 1200,
    height = 800
).configure_axisX(
    labelAngle=0, 
    labelPadding=10, 
    titlePadding=20
).configure_axisY(
    labelPadding=10, 
    titlePadding=20,
).configure_title(
    fontSize=24
).configure_axis(
    labelFontSize=14,
    titleFontSize=14
)
```
